```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
excel_path = "/content/SupplyChainEmissionFactorsforUSIndustriesCommodities.xlsx"
df = pd.read_excel(excel_path, sheet_name="2016_Detail_Commodity")
co2_df = df[df['Substance'].str.lower() == 'carbon dioxide']

summary_stats = co2_df['Supply Chain Emission Factors with Margins'].describe()
print("Summary Statistics for CO₂ Emission Factors (with Margins):")
print(summary_stats)

top10_co2 = co2_df.sort_values(
    by='Supply Chain Emission Factors with Margins', ascending=False
).head(10)

print("\nTop 10 Commodities by CO₂ Emissions:")
print(top10_co2[['Commodity Name', 'Supply Chain Emission Factors with Margins']])

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10_co2,
    y='Commodity Name',
    x='Supply Chain Emission Factors with Margins',
    palette='Reds_r'
)
plt.title('Top 10 Commodities by CO₂ Emission Factor (with Margins) – 2016')
plt.xlabel('CO₂ Emission Factor (kg CO₂e / 2018 USD)')
plt.ylabel('Commodity Name')
plt.tight_layout()
plt.show()
```
